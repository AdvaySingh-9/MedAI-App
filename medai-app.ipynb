# 😕 The Physician’s Burden

Modern clinicians operate in an environment of extreme cognitive and time pressure. In routine clinical practice, physicians are required to simultaneously interpret patient histories, laboratory results, imaging studies, clinical notes, and evolving medical guidelines—often within minutes and under high-stakes conditions.

A significant portion of a physician’s time is spent not on direct patient care, but on **information synthesis and documentation**. Tasks such as summarizing lengthy medical reports, extracting clinically relevant abnormalities, reviewing imaging studies, answering guideline-based clinical questions, and drafting structured medical reports contribute substantially to workload and burnout.

Key challenges faced by physicians include:

- **Information overload:** Patient data is fragmented across reports, PDFs, lab results, and imaging, making rapid synthesis difficult.
- **Time constraints:** Limited consultation time forces clinicians to make decisions quickly, increasing the risk of missed details.
- **Documentation burden:** Writing structured medical reports and notes is repetitive, time-consuming, and often performed after clinical hours.
- **Cognitive fatigue:** Continuous interpretation of complex clinical data increases mental load and contributes to diagnostic error and burnout.
- **Inconsistent access to decision support:** Evidence-based guidance and prior clinical knowledge are not always immediately available at the point of care.

These challenges do not stem from a lack of clinical expertise, but from *workflow inefficiencies* and the growing volume and complexity of medical data. As healthcare systems scale and patient loads increase, the gap between available clinical time and required cognitive effort continues to widen.

Addressing the physician’s burden requires tools that can *augment clinical reasoning, accelerate information synthesis, and reduce non-clinical workload*, without replacing clinician judgment or introducing unsafe automation. This unmet need forms the foundation for the development of **MedAI**.

# 🤖 Why AI — and Why HAI-DEF

The challenges faced by physicians are not solely due to a lack of tools, but due to the *complexity, variability, and unstructured nature of medical data*. Clinical information exists across free-text notes, laboratory reports, PDFs, imaging studies, and evolving clinical guidelines. Traditional rule-based systems or static software workflows are insufficient to handle this level of complexity in real-world clinical settings.

### Why Artificial Intelligence?

Artificial Intelligence, particularly modern foundation models, is well-suited to this problem domain because it can:

- *Understand unstructured medical language*, including clinician notes and narrative reports
- *Synthesize information across multiple data sources*, such as labs, reports, and imaging
- *Adapt to diverse clinical scenarios* without rigid rule definitions
- *Assist clinicians in real time*, reducing cognitive load rather than adding to it

However, not all AI models are appropriate for clinical use.

### Why Not Generic AI Models?

General-purpose language models are trained on broad internet-scale data and often lack:
- Deep exposure to *clinical terminology and medical documentation*
- Awareness of *clinical context, uncertainty, and safety boundaries*
- The discipline to avoid hallucinating diagnoses or recommendations in medical settings

In healthcare, incorrect or fabricated information can lead to serious harm. Therefore, the model choice must prioritize *clinical reliability, medical grounding, and safety-aware behavior*.

### Why HAI-DEF Medical Foundation Models?

HAI-DEF medical foundation models are specifically designed for healthcare applications. They are trained and evaluated on medical data and optimized for tasks such as clinical reasoning, report synthesis, document understanding, and medical image interpretation.

HAI-DEF models are particularly suitable for MedAI because they:

- Demonstrate *strong medical language understanding*
- Handle *clinical summaries and structured report generation* more reliably
- Support *medical image interpretation* in a controlled, assistive manner
- Are designed to *augment clinician decision-making*, not replace it

By using HAI-DEF models, MedAI leverages AI where it adds the most value—*accelerating clinical workflows while preserving physician oversight and responsibility*. 

This alignment between the problem domain and the capabilities of HAI-DEF models ensures that AI is used *appropriately, safely, and effectively*, rather than as a generic or superficial addition.

# 💬 Clinical Consultant

### Explanation
**Clinical Consultant** targets the everyday clinical questions physicians encounter during patient care, such as guideline clarification, clinical reasoning support, and concept validation. This feature is designed for knowledge retrieval and reasoning support, not diagnosis or treatment decisions.

**Inputs**:
- Free-text clinical questions from the physician
(e.g., guideline queries, medication considerations, condition overviews)
- Optional clinical context provided by the user (if available)
 
**Model Processing**:
- The input question is passed to MedGemma 1.5 4B IT model, which is optimized for biomedical language and clinical concepts.
- The prompt constrains the model to operate in an educational and assistive role, ensuring responses remain informational and clinically grounded.
- The model reasons using its medical training to generate structured, relevant explanations aligned with standard clinical knowledge.

**Output**:
- A clear, concise clinical explanation
- Focus on guidelines, mechanisms, risks, or considerations
- No direct diagnosis, prescriptions, or medical orders

### Code

````python
def extract_text_from_pdf(pdf_file): # We will also create function to extract text from a .pdf file.
    reader = PdfReader(pdf_file)
    all_text = ""
    # Loop through pages
    for page_num in range(len(reader.pages)): # Read all the pages in PDF.
        page = reader.pages[page_num]
        text = page.extract_text()
        page_number = {page_num+1}
        if text:
            all_text += text + "\n"
    return all_text.strip()   #  # Return extracted text from the PDF

def chat(user_input, pdf_file): # user_input is the question asked by the user
    if not user_input and not pdf_file:
        return "Please enter a question or upload a medical document."

    pdf = ""
    
    if pdf_file:
        pdf = extract_text_from_pdf(pdf_file)
    
    prompt = f"""You are a clinical decision support AI for doctors.
Provide a detailed answer to doctors question in plain text.
Do not give any treatment for any disease.
If context is given, then use that to answer.
Context:
{pdf}

Doctor’s Question:
{user_input}

Answer:.
"""
# Use :. after the prompt

    if model.device == "cpu":                            # If the app is running on CPU, then set the text limit to 512 for faster response 
        max_new_tokens = 512
    else:
        max_new_tokens = 1024                            # If the app is NOT running on CPU, then set the text limit to 1024

    inputs = processor(text=prompt,                      # Give input to the model
                   return_tensors="pt"
                  ).to(model.device)

    
    outputs = model.generate(**inputs,                   # Generate output
                        max_new_tokens=max_new_tokens,
                        do_sample=False,
                        repetition_penalty=1.2,
                        )

    response = processor.decode(outputs[0],             # Decode text from a model’s output.
                            skip_special_tokens=True
                           )
    # Response might repeat the prompt, or give a JSON formatted output or the model might check it's own answers in it's response
    # To avoide this noise, we will remove all the text before ":." (prompt). And also remove all the text after "```json, ---,Critique"
    response = response.partition(":.")[2].partition("```json")[0].partition("---")[0].partition("Critique")[0] 
    
    return response

print("✅ Chat function created!")
````

- 🧠 Lowers cognitive load: Supports clinical reasoning during complex cases.
- 📚 Improves consistency: Encourages guideline-aligned thinking.
- ⚠️ Reduces risk: Helps avoid errors caused by outdated knowledge or uncertainty.

# 📄 Clinical Synthesis

### Explanation
**Clinical Synthesis** targets one of the biggest pain points in clinical practice:

*extracting clinically meaningful insights from long, unstructured medical documents* such as reports, discharge summaries, or PDF-based records.

This feature is designed to **analyze and reorganize medical text**, not just shorten it.

**Inputs**:
- Large medical reports or clinical documents (plain text or PDF-extracted text)
- Optional doctor’s question related to the document
(e.g., “What are the major red flags?” or “What needs urgent follow-up?”)

**Model Processing**:
- The document is passed to MedGemam model capable of long-context clinical reasoning.
- The model parses the text to identify:
  - Key clinical findings
  - Abnormal values with clinical relevance
  - Potential risks or red flags
  - Clinical considerations for further evaluation
- If a doctor’s question is present, the model *grounds its answer strictly in the provided document*, avoiding hallucinations.
- Output is structured using predefined clinical headings to ensure consistency and readability.

**Output**:
- A structured clinical summary with clearly separated sections:
  - Key clinical findings
  - Abnormal values
  - Risks / red flags
  - Clinical considerations
- Context-aware answers to physician questions (if provided)
- No new diagnoses or treatment decisions are introduced

### Code

````python
def summarise(text_input, pdf_file): # Takes user's question (text_input) or PDF file (pdf_file)
    if pdf_file is not None:
        report_text = extract_text_from_pdf(pdf_file)
    elif text_input.strip() != "":
        report_text = text_input
    else:
        return "⚠️ Please paste a report or upload a PDF."

    
    prompt = f"""You are a clinical decision support AI for licensed doctors.

Analyze the following medical report and provide:
1. Key clinical findings
2. Abnormal values with clinical relevance
3. Potential risks or red flags
4. Clinical considerations for further evaluation

Do not provide diagnoses or treatment instructions.
Use a professional medical tone.

Report:
{report_text}

Answer:.
"""
# Use :. after the prompt

    
    if model.device == "cpu":                            # If the app is running on CPU, then set the text limit to 512 for faster response 
        max_new_tokens = 512
    else:
        max_new_tokens = 1024                            # If the app is NOT running on CPU, then set the text limit to 1024
    
    
    inputs = processor(text=prompt,                      # Give input to the model
                   return_tensors="pt"
                  ).to(model.device)

    
    outputs = model.generate(**inputs,                   # Generate output
                        max_new_tokens=1024,
                        do_sample=False,
                        repetition_penalty=1.2,
                        )

    response = processor.decode(outputs[0],             # Decode text from a model’s output.
                            skip_special_tokens=True
                           )
# Response might repeat the prompt or give useless text after --- (ending line)
# To avoide this noise, we will remove all the text before ":." (prompt) and after "---".
    response = response.partition(":.")[2].partition("---")[0]
    return response

print("✅ Summarisation function created!")
````

### Clinical Impact
- 📉 Reduces information overload: Converts long reports into actionable clinical insight.
- ⏱️ Saves time: Doctors no longer need to manually scan entire documents.
- 🚨 Improves patient safety: Highlights abnormal values and red flags that may be overlooked.
- 🧩 Supports continuity of care: Helps clinicians quickly understand unfamiliar patient histories.
- 🧠 Enhances decision readiness: Physicians receive organized information aligned with real clinical thinking.

# 🖼️ Clinical Imaging

### Explanation
**Clinical Imaging** supports physicians in understanding and reviewing medical images within a clinical context. This feature is intended to assist with visual interpretation and structured observation, not to replace formal radiology reporting or make definitive diagnoses.

**Inputs**:
- Medical images provided by the physician
(e.g., chest X-rays, CT scans, MRI images)
- Optional free-text clinical question related to the image
(e.g., “Are there signs of consolidation?” or “Is cardiomegaly visible?”)

**Model Processing**:
- The provided image is processed by a MedGemma capable of analyzing both visual and clinical information.
- If a clinical question is present, the model focuses its analysis on the specific region or concern mentioned.
- If no question is provided, the model generates a neutral, structured description of visible findings using standard medical terminology.
- The prompt explicitly restricts the model to observational and descriptive analysis, avoiding definitive diagnoses or treatment recommendations.

**Output**:
- Structured observations of visible imaging findings
- Clear, clinically relevant descriptions suitable for clinical review
- Cautious language emphasizing correlation with clinical context and further evaluation where appropriate

### Code

````python
def review_image(image_file, doctor_question): # takes an image and a question from the user
    if image_file is None:
        return "⚠️ Please upload an image."

    # Load image
    image = Image.open(image_file.name).convert("RGB")     # open the image

    # Build conversation
    conversation = [
        {
            "role": "user",
            "content": [
                {"type": "image"},
                {
                    "type": "text",
                    "text": f"""
You are a medical image analysis assistant designed to support doctors.
You must NOT provide a definitive diagnosis or claim certainty.
Your role is to carefully describe visual findings and highlight areas that appear suspicious.

GENERAL RULES:
- Always produce a LONG, detailed, and structured response.
- Use professional medical language, but explain it in simple terms.
- Never say “this confirms” or “this is definitely”.
- Use cautious wording such as:
  “may suggest”, “could indicate”, “appears suspicious”, “cannot be ruled out”.
- If no abnormality is visible, clearly state that.

IMAGE HANDLING:
1. Analyze the entire image before responding.
2. Identify visible organs, structures, and anatomical regions.
3. Look for asymmetry, opacities, shadows, lesions, or unusual patterns.

QUESTION HANDLING:
- If the doctor asks a question:
  - Answer it in detail first.
  - Then explain any relevant visual findings from the image.
- If no question is provided:
  - Give a detailed descriptive report of the image and its visible structures.

SUSPECTED AREA DESCRIPTION:
- If any area appears suspicious:
  - Describe its location ONLY using human-readable spatial terms:
    - upper left
    - upper right
    - lower left
    - lower right
    - central region
  - Mention the organ or structure involved.
  - Briefly explain why the area appears suspicious (visual reasoning only).

IMPORTANT:
- Do NOT return coordinates, numbers, or bounding boxes.
- Do NOT draw conclusions or diagnoses.
- Do NOT recommend treatment.

OUTPUT STRUCTURE (STRICT):
1. Detailed Explanation
2. Suspected Area Location (only if applicable)
3. Clinical Caution Note

Always end with:
“These findings are observational and should be correlated clinically.”

Dopctor's question:
{doctor_question}

Answer:.
"""
                }
            ]
        }
    ]

    prompt = processor.apply_chat_template(             # creating the prompt with image and text
        conversation,
        add_generation_prompt=True
    )


    if model.device == "cpu":                            # If the app is running on CPU, then set the text limit to 512 for faster response 
        max_new_tokens = 512
    else:
        max_new_tokens = 1024                            # If the app is NOT running on CPU, then set the text limit to 1024
    
    
    inputs = processor(                      # Give input to the model
        images=image,
        text=prompt,
        return_tensors="pt"
    ).to(model.device)

    outputs = model.generate(                   # Generate output
        **inputs,
        max_new_tokens=1024,
        do_sample=True,
        temperature=0.5,
        top_p=0.9,
        repetition_penalty=1.2
    )

    response = processor.decode(
        outputs[0],
        skip_special_tokens=True
    ).strip()

    # Response might repeat the prompt
    # To avoide this noise, we will remove all the text before ":." (prompt) and replace "model" word from the output
    response = response.partition(":.")[2].replace("model", "")
    return response
print("✅ Medical Image Review function created!")
````

### Clinical Impact
- Assists clinicians in quickly extracting key visual information from medical images.
- Supports structured and consistent interpretation of imaging findings.
- Reduces cognitive load during busy clinical workflows.
- Encourages safe, observation-based analysis rather than premature diagnostic conclusions.

# 📝 Clinical Scribe

### Explanation
**Clinical Scribe** assists physicians in drafting structured medical reports from minimal and loosely written clinical inputs. This feature is designed to reduce documentation burden by transforming rough notes into a coherent clinical draft, without attempting to finalize or validate medical decisions.

**Inputs**
- Basic patient information
(e.g., age, gender, presenting complaint)
- Brief or unstructured clinical notes
(e.g., shorthand observations, incomplete sentences, bullet points)
- Optional medical history (if available)

**Model Processing**
- The input data is provided to MedGemma trained on clinical language and documentation patterns.
- The prompt guides the model to organize the information into a standard medical report structure (e.g., patient information, presenting complaint, examination, investigations, assessment).
- The model focuses on clarity, clinical coherence, and professional tone, while explicitly avoiding the introduction of new medical facts, diagnoses, or treatment decisions not present in the input.

**Output**
- A structured draft medical report
- Clinically appropriate language and formatting
- Clearly organized sections suitable for physician review and editing
- A non-final document intended for completion and validation by a clinician

````python
def report_creator(patient_info, clinical_notes, pdf_file, image_file, medical_history): # takes patient info (like name, age, gender), clinical notes, PDF, Image and medical history
    
    if not patient_info or not clinical_notes:   # patient info and clinical notes is required for generating medical report
        return "⚠️ Please provide patient information and clinical notes."
    image = None

    pdf_text = ""
    if pdf_file is not None:
        pdf_text = extract_text_from_pdf(pdf_file)

    content = []

    if image is not None:
        content.append({"type": "image"})

    content.append({
        "type": "text",
        "text": f"""
You are a professional Clinical Documentation Assistant. 
Generate a formal Mecial Report using professional formatting.

STRUCTURE RULES:
- Use this exact format in generating a medical report:-
Patient Information
* Name: [Insert Name]
* Age: [Insert Age]
* Gender: [Insert Gender]
* Date: [Insert Date]
Medical History
* Past illnesses: [Insert details]
* Surgeries: [Insert details]
* Allergies: [Insert details]
* Medications: [Insert details]
Presenting Complaint
* Symptoms: [Insert details]
* Duration: [Insert details]
Examination Findings
* Vital signs: BP [ ], HR [ ], Temp [ ]
* Physical exam: [Insert findings. You can also insert what you find]
Investigations
* Blood tests: [Insert results]
* Imaging: [Insert results. You can also insert what you find]
Diagnosis
* Primary condition: [Insert diagnosis]
* Secondary conditions: [Insert if any]
Treatment Plan
* Medications prescribed: [Insert details]
* Lifestyle advice: [Insert details]
* Follow-up: [Insert timeline]
- You can skip any detail but only if that is not given.

DATA:
Patient Information: {patient_info}
Clinical Notes: {clinical_notes}
Medical History: {medical_history if medical_history!="" else "No Meedica History Provide."}
Context: {pdf_text if pdf_text else "No additional context provided."}

Generate the Report now:.
"""
    })

    conversation = [{"role": "user", "content": content}]

    prompt = processor.apply_chat_template(             # creating the prompt with image and text
        conversation,
        add_generation_prompt=True
    )


    if model.device == "cpu":                            # If the app is running on CPU, then set the text limit to 512 for faster response 
        max_new_tokens = 512
    else:
        max_new_tokens = 1024                            # If the app is NOT running on CPU, then set the text limit to 1024

    
    inputs = processor(                      # Give input to the model
        images=image,
        text=prompt,
        return_tensors="pt"
    ).to(model.device)

    outputs = model.generate(                   # Generate output
        **inputs,
        max_new_tokens=1024,
        do_sample=False,
        repetition_penalty=1.1
    )

    response = processor.decode(outputs[0],
                            skip_special_tokens=True
                           )

    # Response might repeat the prompt
    # To avoide this noise, we will remove all the text before ":." (prompt) and replace "model" word from the output
    response = response.partition(":.")[2].replace("model", "")
    
    return response
print("✅ Report Creator function created!")
````

### Clinical Impact
- Reduces time spent converting handwritten or rough notes into formal reports
- Improves consistency and clarity of clinical documentation
- Helps prevent omission of key findings during manual transcription
- Supports clinicians in high-workload environments while maintaining full clinical control

  - Medical reports or document text

This interface is intended as a *demonstration and evaluation tool*, showcasing how the system could integrate into clinical workflows. It is *not a diagnostic system*, nor a replacement for electronic health record (EHR) platforms.

# ⚠️ Safety Considerations and Limitations
MedAI is designed as a clinical decision-support and documentation assistance tool, not as a diagnostic or treatment system. The following considerations and limitations are critical to ensure safe, ethical, and responsible use in real-world clinical environments.

### 1. Non-Diagnostic and Non-Prescriptive Use
  - MedAI does not provide diagnoses, treatment plans, or prescriptions.
  - All outputs are informational, educational, or assistive in nature.
  - Final clinical decisions remain entirely under the responsibility of licensed healthcare professionals.

### 2. Human-in-the-Loop Requirement
  - The application is explicitly designed to operate with active physician oversight.
  - Model outputs should be reviewed, validated, and contextualized by clinicians before any clinical action is taken.
  - MedAI supports, but does not replace, clinical judgment.
### 3. Risk of Model Hallucinations
  - Like all large language models, MedAI may generate:
    - Incomplete information
    - Over-generalized responses
    - Confident but incorrect statements
  - Prompts are constrained to reduce this risk, but hallucinations cannot be fully eliminated.
  - Users are expected to verify critical information against trusted clinical guidelines and institutional protocols.
### 4. Data Privacy and Patient Safety
  - The system is designed to operate without requiring persistent storage of patient data.
  - No patient-identifiable information should be entered unless compliant with applicable privacy regulations.
  - This notebook demonstrates technical feasibility and does not represent a production-hardened, regulatory-approved system.

### 5. Model Knowledge and Temporal Limitations
  - The underlying medical foundation models rely on static training data.
  - The model may not reflect:
    - The most recent clinical guidelines
    - Emerging therapies
    - Institution-specific protocols
  - Clinicians must rely on up-to-date sources for time-sensitive decisions.

### 6. Bias and Generalization Risks
  - Medical AI models may reflect biases present in training data.
  - Performance may vary across:
    - Patient populations
    - Rare conditions
    - Edge cases outside common clinical patterns
  - The system is not validated for universal clinical generalization.

### 7. Technical and Deployment Constraints
  - Performance depends on available hardware resources.
  - CPU-only environments may experience:
    - Increased latency
    - Reduced response length
  - Edge deployment requires:
    - Model compression
    - Quantization
    - Runtime adaptation beyond this prototype

### 8. Intended Scope of Use
  - MedAI is intended for:
    - Clinical education
    - Documentation assistance
    - Workflow support
    - Knowledge referencing
  - It is not intended for autonomous clinical decision-making or emergency care scenarios.
